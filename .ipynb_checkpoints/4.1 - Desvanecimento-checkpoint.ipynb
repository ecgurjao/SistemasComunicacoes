
# Desvanecimento em Sistemas de Comunicações

O desvanecimento (*fading*) é um fenômeno causado principalmente pela propagação multipercurso, mobilidade dos terminais, obstáculos, reflexão, difração e espalhamento do sinal transmitido.


## Definição

Em um sistema de comunicações, o **desvanecimento** é a variação aleatória da amplitude, fase ou potência do sinal recebido ao longo do tempo, frequência ou espaço.


Em um canal com desvanecimento, o sinal recebido pode ser modelado por:
$$
r(t)=h(t)s(t)+n(t)
$$
sendo:
- $h(t)$ é a resposta ao impulso do canal;
- $s(t)$ é o sinal transmitido;
- $n(t)$ é o ruído.

O termo $h(t)$ representa os efeitos do canal, incluindo atenuação, variação de fase e propagação multipercurso.



## Tipos de Desvanecimento

### Desvanecimento de Grande Escala

Ocorre devido a variações lentas na potência média recebida. Está associado a:
- distância entre transmissor e receptor;
- obstáculos grandes;
- sombreamento por prédios, árvores ou relevo;
- perdas de percurso (*path loss*).

> Exemplo: (Equação de Friis)
$$
P_r(d) = \frac{P_t G_tG_r\lambda^2}{(4\pi)^2d^\alpha L}
$$
sendo:
- $P_r(d)$ é a potência recebida a uma distância $d$;
- $P_t$ a potência transmitida;
- $G_t$ ganho da antena transmissora;
- $G_r$ ganho da antena receptora;
- $\alpha$ é o expoente de perda de percurso. No espaço livre $\alpha = 2$.
- $L$ perdas do sistema não relacionadas à propagação.

> Exemplo:

> Se um transmissor produz 50 W de potência, expresse a potência de transmissão em unidades de (a) dBm, e (b) dBW. Se 50 W forem aplicados a uma antena de ganho unitário com frequência de portadora de 900 MHz, ache a potência recebida em dBm a uma distância livre de 100 m da antena. Qual é a $P_r(10\ \text{km})$? Considere um ganho unitário para a antena receptora.

> **Solução:**


> Potência de transmissão $P_t = 50\ \text{W}$

> Frequência da portadora $f_c = 900\ \text{MHz}$

> Potência do transmissor:
$$P_t(\text{dBm})
 =
 10\log\left[\frac{P_t(\text{mW})}{1\ \text{mW}}\right]
=
10\log[50 \times 10^3]
=
47,0\ \text{dBm}
$$

> Potência do transmissor,
$$
P_t(\text{dBW})
=
10\log\left[\frac{P_t(\text{W})}{1\ \text{W}}\right]
=
10\log[50]
=
17,0\ \text{dBW}
$$

> Potência recebida:
$$
P_r
=
\frac{P_t G_t G_r \lambda^2}
{(4\pi)^2 d^2 L}
$$
> Substituindo os valores:
$$
P_r
=
\frac{
50(1)(1)\left(\frac{1}{3}\right)^2
}
{
(4\pi)^2 (100)^2 (1)
}
$$

$$
=
3,5 \times 10^{-6}\ \text{W}
=
3,5 \times 10^{-3}\ \text{mW}
$$

> Convertendo para dBm:
$$
P_r(\text{dBm})
=
10\log P_r(\text{mW})
=
10\log(3,5 \times 10^{-3})
=
-24,5\ \text{dBm}
$$

> A potência recebida a 10 km: $P_r = -64,5\ \text{dBm}.$

## Mecanismos de Propagação

### Reflexão

Quando uma onda colide com um meio cujas dimensões são muito maiores do que o comprimento de onda, e parte da energia retorna ao meio original.


> Exemplo: Canal com dois raios:
$$ r(t) = s(t) + \alpha s(t - T_0)$$
Na frequência
$$
R(\omega) = S(\omega) + \alpha S(\omega)e^{-j\omega T_0} = S(\omega) (1 + \alpha e^{-j\omega T_0})$$
e
$$
H(\omega) = \frac{R(\omega)}{S(\omega)} = 1 + e^{-j\omega T_0}.
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
T0 = 1
w = np.linspace(-4*np.pi, 4*np.pi, 2000)

# Função de resposta em frequência
H = 1 + np.exp(-1j * w * T0)

# Módulo e fase
modulo = np.abs(H)
#fase = np.angle(H)
# Comentar a linha anterior e descomentar a linha abaixo para evitar os saltos em \pi
fase = np.unwrap(np.angle(H))

# Gráfico do módulo
plt.figure(figsize=(10, 4))
plt.plot(w, modulo)
plt.xlabel(r'$\omega$ rad/s')
plt.ylabel(r'$|H(\omega)|$')
plt.title(r'Módulo de $H(\omega)=1+e^{-j\omega T_0}$')
plt.grid(True)
plt.show()

# Gráfico da fase
plt.figure(figsize=(10, 4))
plt.plot(w, fase)
plt.xlabel(r'$\omega$ rad/s')
plt.ylabel(r'$\angle H(\omega)$ rad')
plt.title(r'Fase de $H(\omega)=1+e^{-j\omega T_0}$')
plt.grid(True)
plt.show()
```

### Difração

Ocorre quando a onda encontra um obstáculo ou borda e consegue contorná-lo parcialmente.

Esse fenômeno permite que sinais sejam recebidos mesmo sem linha de visada direta (*Non-Line of Sight – NLOS*).

> Exemplo: uma estação de rádio é recebida sem que haja visada com a antena tranmissora.



### Dispersão

Ocorre quando a onda encontra objetos menores ou comparáveis ao comprimento de onda, fazendo com que a energia seja espalhada em várias direções.

> Exemplo: ondas milimétricas (e.g 5G mmWave).

### Desvanecimento de Pequena Escala

Ocorre devido à interferência construtiva e destrutiva entre múltiplas cópias do sinal que chegam ao receptor por caminhos diferentes.

Esse tipo de desvanecimento pode causar variações rápidas no sinal recebido, mesmo para pequenos deslocamentos do receptor.

Principais causas:
- multipercurso;
- mobilidade;
- efeito Doppler;
- atraso entre componentes do sinal recebido.


## Classificação do Desvanecimento

O desvanecimento pode ser classificado de acordo com diferentes critérios.


## Quanto à Seletividade em Frequência

### Desvanecimento Plano (*Flat Fading*)

Ocorre quando todas as componentes de frequência do sinal são afetadas aproximadamente da mesma forma.

Condição típica:
$$$
sendo:
- $B_s$ é a largura de banda do sinal;
- $B_c$ é a largura de banda de coerência do canal.

Nesse caso, o canal altera principalmente a amplitude e a fase do sinal.


### Desvanecimento Seletivo em Frequência

Ocorre quando diferentes componentes espectrais do sinal são afetadas de maneiras diferentes.

Condição típica:
$$
B_s > B_c
$$

Esse tipo de desvanecimento pode causar **interferência intersimbólica** (*ISI*), pois diferentes versões atrasadas do sinal se sobrepõem no receptor.


## Quanto à Variação Temporal

### Desvanecimento Lento (*Slow Fading*)

O canal varia lentamente em relação à duração do símbolo.

Condição típica:
$$
T_s \ll T_c
$$
sendo:
- $T_s$ é o tempo de símbolo;
- $T_c$ é o tempo de coerência do canal.


### Desvanecimento Rápido (*Fast Fading*)

O canal varia rapidamente durante a transmissão de um símbolo ou de um pequeno conjunto de símbolos.

Condição típica:
$$
T_s \geq T_c
$$

Esse tipo de desvanecimento está associado à mobilidade e ao efeito Doppler.


## Distribuições de Probabilidade Associadas ao Desvanecimento

Como o desvanecimento possui natureza aleatória, é comum modelar a amplitude do canal por distribuições de probabilidade.


## Distribuição Rayleigh

A distribuição Rayleigh é utilizada quando não existe uma componente dominante em linha de visada (*Line of Sight – LOS*).

É típica em ambientes urbanos densos, onde o sinal recebido resulta da soma de várias componentes refletidas e espalhadas.

A envoltória $R$ do sinal recebido pode ser modelada por:
$$
f_R(r)=\frac{r}{\sigma^2}e^{-\frac{r^2}{2\sigma^2}}, \quad r \geq 0
$$

## Distribuição Rician

A distribuição Rician é utilizada quando existe uma componente dominante, geralmente associada à linha de visada.

Nesse caso, o sinal recebido é composto por:
- uma componente direta dominante;
- várias componentes refletidas ou espalhadas.

A intensidade da componente dominante é representada pelo fator $K$:
$$
K=\frac{\text{potência da componente dominante}}{\text{potência das componentes espalhadas}}
$$

Quanto maior o valor de $K$, mais forte é a componente direta.



## Distribuição Nakagami-$m$

A distribuição Nakagami-$m$ é um modelo flexível que pode representar diferentes níveis de severidade do desvanecimento.

O parâmetro $m$ controla a intensidade do desvanecimento:
- $m < 1$: desvanecimento severo;
- $m = 1$: caso equivalente ao Rayleigh;
- $m > 1$: desvanecimento menos severo.

É bastante utilizada para modelagem empírica de canais sem fio.

## Sombreamento Log-normal

O sombreamento de grande escala é frequentemente modelado por uma distribuição log-normal.

Em dB, a potência recebida pode ser escrita como:
$$
P_r(dB)=\overline{P_r}(dB)+X_\sigma
$$
sendo:
- $\overline{P_r}(dB)$ é a potência média;
- $X_\sigma$ é uma variável aleatória Gaussiana de média zero;
- $\sigma$ representa o desvio padrão do sombreamento.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh, rice, nakagami

# Parâmetros
N = 100000
r = np.linspace(0, 5, 1000)

# Distribuições teóricas
pdf_rayleigh = rayleigh(scale=1).pdf(r)

# Rician: parâmetro b representa a intensidade relativa da componente dominante
pdf_rician_K_baixo = rice(b=1, scale=1).pdf(r)
pdf_rician_K_alto = rice(b=4, scale=1).pdf(r)

# Nakagami
pdf_nakagami_m05 = nakagami(nu=0.5, scale=1).pdf(r)
pdf_nakagami_m1 = nakagami(nu=1.0, scale=1).pdf(r)
pdf_nakagami_m3 = nakagami(nu=3.0, scale=1).pdf(r)

plt.figure(figsize=(10, 5))
plt.plot(r, pdf_rayleigh, label="Rayleigh")
plt.plot(r, pdf_rician_K_baixo, label="Rician - componente dominante fraca")
plt.plot(r, pdf_rician_K_alto, label="Rician - componente dominante forte")
plt.xlabel("Amplitude normalizada")
plt.ylabel("Densidade de probabilidade")
plt.title("Distribuições associadas ao desvanecimento")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(r, pdf_nakagami_m05, label="Nakagami m=0.5")
plt.plot(r, pdf_nakagami_m1, label="Nakagami m=1")
plt.plot(r, pdf_nakagami_m3, label="Nakagami m=3")
plt.xlabel("Amplitude normalizada")
plt.ylabel("Densidade de probabilidade")
plt.title("Distribuição Nakagami-m para diferentes valores de m")
plt.grid(True)
plt.legend()
plt.show()
```

## Efeitos do Desvanecimento no Sinal Recebido

O desvanecimento pode afetar o sinal recebido de várias formas:

### Variação da Amplitude

A amplitude do sinal pode variar rapidamente, causando quedas profundas de potência.

$$
r(t)=a(t)s(t)
$$

onde $a(t)$ representa a envoltória aleatória do canal.


### Variação de Fase

Além da amplitude, o canal pode introduzir variações de fase:
$$r
(t)=a(t)e^{j\phi(t)}s(t)
$$


### Interferência Intersimbólica

Quando o canal é seletivo em frequência, diferentes versões atrasadas do sinal podem se sobrepor:
$$
r(t)=\sum_{k} h_k s(t-\tau_k)+n(t)
$$

Esse efeito causa distorção temporal e pode prejudicar a detecção dos símbolos.

---

### Redução da SNR Instantânea

A SNR recebida passa a variar aleatoriamente:
$$
\gamma(t)=\frac{|h(t)|^2P_s}{N_0B}
$$
sendo:
- $\gamma(t)$ é a SNR instantânea;
- $P_s$ é a potência do sinal;
- $N_0B$ representa a potência de ruído na banda;
- \(|h(t)|^2\) representa o ganho instantâneo do canal.

```python

# Exemplo: sinal senoidal recebido sob desvanecimento Rayleigh

fs = 2000
T = 1
t = np.arange(0, T, 1/fs)

fc = 50
s = np.cos(2*np.pi*fc*t)

# Envoltória Rayleigh variando lentamente
np.random.seed(10)
num_blocos = 40
amostras_por_bloco = len(t) // num_blocos

envoltoria_blocos = rayleigh(scale=1).rvs(size=num_blocos)
envoltoria = np.repeat(envoltoria_blocos, amostras_por_bloco)

# Ajuste de tamanho
envoltoria = envoltoria[:len(t)]

# Sinal recebido com fading
r = envoltoria * s

plt.figure(figsize=(12, 4))
plt.plot(t, s, label="Sinal transmitido")
plt.plot(t, r, label="Sinal recebido com desvanecimento", alpha=0.8)
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.title("Efeito do desvanecimento Rayleigh no sinal recebido")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(t, envoltoria)
plt.xlabel("Tempo (s)")
plt.ylabel("Ganho do canal")
plt.title("Envoltória do canal com desvanecimento")
plt.grid(True)
plt.show()
```

```python

# Exemplo: impacto do fading na SNR instantânea

snr_media_db = 20
snr_media_linear = 10**(snr_media_db/10)

# Ganho de canal Rayleigh: |h|^2 tem distribuição exponencial
h = (np.random.randn(N) + 1j*np.random.randn(N)) / np.sqrt(2)
ganho_canal = np.abs(h)**2

snr_instantanea = ganho_canal * snr_media_linear
snr_instantanea_db = 10*np.log10(snr_instantanea + 1e-12)

plt.figure(figsize=(10, 5))
plt.hist(snr_instantanea_db, bins=100, density=True)
plt.xlabel("SNR instantânea (dB)")
plt.ylabel("Densidade")
plt.title("Distribuição da SNR instantânea em canal Rayleigh")
plt.grid(True)
plt.show()

print(f"SNR média considerada: {snr_media_db} dB")
print(f"SNR instantânea mínima observada: {np.min(snr_instantanea_db):.2f} dB")
print(f"SNR instantânea média observada: {10*np.log10(np.mean(snr_instantanea)):.2f} dB")
```

## Consequências Práticas em Sistemas de Comunicações

O desvanecimento pode provocar:
- queda momentânea da potência recebida;
- aumento da taxa de erro de bit;
- perda de pacotes;
- falhas de sincronismo;
- variações na qualidade do enlace;
- redução da cobertura;
- necessidade de retransmissão;
- redução da taxa de transmissão adaptativa.

Em sistemas reais, como Wi-Fi, 4G, 5G, enlaces via rádio e comunicações via satélite, técnicas são utilizadas para reduzir seus efeitos.


## Técnicas de Mitigação

Algumas técnicas comuns são:
- diversidade espacial com múltiplas antenas;
- diversidade em frequência;
- diversidade temporal;
- equalização;
- OFDM;
- códigos corretores de erro;
- interleaving;
- controle adaptativo de modulação e codificação;
- MIMO;
- beamforming.
