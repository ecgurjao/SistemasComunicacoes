# Transformada de Hilbert

Dado um sinal $x(t)$ essa transformada é dada por
$$
x_h(t) = \frac{1}{2\pi}\int \frac{x(\tau)}{t - \tau}d\tau
$$
que ainda pode ser vista como a convolução de $x(t)$ com $h(t) = \frac{1}{\pi t}$.

Na frequência
$$
X_h(\omega) = -jsign(\omega)X(\omega)
$$
sendo a função sinal dada por
$$
    sign(\omega) = \left\{\begin{array}{c} 1, \omega > 0 \\ 0, \omega = 0, \\ -1, \omega < 0 \end{array}\right.
$$

Observe que na frequência a transformada de Hilbert não altera o módulo, porém, frequências positivas sofrem um atraso de $90^\circ$ e frequências negativas um avnaço de $90^\circ$.
